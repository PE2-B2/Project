## 3. How to use this program
1. First, you need to add the data files that you want to analyze in the data folder and run the "run.py". If you do so, you can see a pop-up window below.



2. Second, you can set the wafer, coordinates, and device ID in the pop up window.
    - Wafer : The input value must be the title of the each data file(e.g., D07). 
    - Coordinates : The input value must be the coordinates of wafer(e.g.,(1,1)), or if you want to check the entire coordinates, the input can be 'all'.
    - Device ID : Since this program only provides analysis of LMZ files, the input value must be 'LMZ'.
3. Third, if you click the 'Show Output', the results pop-up window will come up, and if you click the 'Save Output', the results will be saved in res folder.
4. Lastly, after setting up, you can run the program easily by clicking the 'OK'. Also, you can comfortably check the results saved in the res folder by clicking the 'Result folder'. Other values, such as time for fitting degree 7, R square value for degree 7, Max and Min value, can be checked through python result window.
