## Visualize analyzed data

### Introduction

This file is about comparision analysis for each wafer of sample data.  
Our sample data have four wafer, and we analyzed about device name 'LMZ'.  
At now, our team have a csv file result, so we extracted some parts of the data to compare.  

Our team used the library named 'seaborn' to visualize analyzed data.  
Darkgrid was applied to all graphs to have better visibility.

#### Import some library and bring our result about sample data processing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Bring the result of sample data

```python
df_train = pd.read_csv('../res/csvRes/analyzedResult.csv')
```

### 1. ErrorFlag

This is a graph abour 'ErrorFlag for rsq of I-v'.  
Our team used the errorflag for I-V graph fitting.

```python
sns.swarmplot(x = 'Wafer', y = 'ErrorFlag', data = df_train)
sns.set_style("darkgrid")
plt.title('ErrorFlag for rsq of I-V', fontsize = 18)
plt.xlabel('WaferID', fontsize = 13)
plt.ylabel('ErrorFlag', fontsize = 13)
```

### 2. Analysis Wavelength

This is a graph about analysis wavelength for each wafer.  
You can compare the value of wavelength in each wafer.  

```python
sns.swarmplot(x = 'Wafer', y = 'Analysis Wavelength', data = df_train)
sns.set_style("darkgrid")
plt.title('Analysis Wavelength', fontsize = 18)
plt.xlabel('WaferID', fontsize = 13)
plt.ylabel('Wavelength [nm]', fontsize = 13)
```

### 3. R square of reference spectrum data

These are two graph about R square value of reference spectrum data for each wafer.  
Left graph used 'swarmplot' for each result, and right graph used 'boxplot' for whole result of each wafer.

```python
fig = plt.figure(figsize = (15, 5))
plt.suptitle('Rsq of Ref.spectrum comparison', fontsize=20)
sns.set_style("darkgrid")

pic_1 = plt.subplot(1, 2, 1)
sns.swarmplot(x = 'Wafer', y = 'Rsq of Ref. spectrum', data = df_train, ax = pic_1)
plt.xlabel('Wafer ID', fontsize=13)
plt.ylabel('Rsq of ref. spec.', fontsize=13)

pic_2 = plt.subplot(1, 2, 2)
sns.boxplot(x = 'Wafer', y = 'Rsq of Ref. spectrum', data = df_train, ax = pic_2)
plt.xlabel('Wafer ID', fontsize=13)
plt.ylabel('Rsq of ref. spec.', fontsize=13)
```

### 4. Max transmission of reference spectrum data

These are two graph for compare maximum value of refernce spectrum data about each wafer.  
Left graph used 'swarmplot' for each maximum value, and right graph used 'boxplot' for whole result of each wafer.

```python
fig = plt.figure(figsize = (15, 5))
plt.suptitle('Max transmission of Ref spec(dB) comparison', fontsize = 20)
sns.set_style("darkgrid")


pic_1 = plt.subplot(1, 2, 1)
sns.swarmplot(x = 'Wafer', y = 'Max transmission of Ref. spec. (dB)', data = df_train, ax = pic_1)
plt.xlabel('Wafer ID', fontsize = 13)
plt.ylabel('Max transmission of ref. spec.', fontsize = 13)

pic_2 = plt.subplot(1, 2, 2)
sns.boxplot(x = 'Wafer', y = 'Max transmission of Ref. spec. (dB)', data = df_train, ax = pic_2)
plt.xlabel('Wafer ID', fontsize = 13)
plt.ylabel('Max transmission of ref. spec.', fontsize = 13)
```
